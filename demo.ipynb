```python
import pypistats
import pandas as pd
from pandas import DataFrame, Series
pd.options.plotting.backend = "plotly"
```

```python
def ser_for(pkg: str) -> Series:
    """pkg downloads of tbe past 180 days in time series"""
    df = pypistats.overall(pkg, format="pandas", mirrors=True, total='daily')
    df = df[['date', 'downloads']].dropna().set_index('date').sort_index()
    df.index = pd.to_datetime(df.index)
    return df['downloads']


def combined_pkgs(names: list[str]) -> DataFrame:
    """time series with pkg downloads in column"""
    dfs = [ser_for(pkg) for pkg in names]
    df = pd.concat(dfs, axis='columns')
    df.columns = names
    return df
```

```python
combined_pkgs(['sanic', 'django', 'fastapi']).plot()
```
